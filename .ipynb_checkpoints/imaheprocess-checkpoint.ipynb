```python
import cv2

# Load the video file
source = cv2.VideoCapture(r'C:\Users\prave\Desktop\python-lening\v1.mp4')

# Read the first frame
ref, img = source.read()
print(ref)

# Check if the frame was read successfully
if not ref:
    print("Error: Couldn't read the video file or video file is empty.")
else:
    while(True):
         cv2.imshow('IMG', img)
         cv2.waitKey(1)  # Corrected here
    cv2.destroyAllWindows()
    
   
```
